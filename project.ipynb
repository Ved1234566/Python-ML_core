```python
import pandas as pd
```

```python
df = pd.read_csv('Attrition.csv')
```

```python
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default = "plotly_white"
```

```python
print(df.isnull().sum())
```

```python
# CHECK PERCENTAGE OF ATTRITION BY DEAPRTMENT
# FILTER THE DATA TO SHOW ONLY "YES" VALUES IN THE "ATTRITION" COLUMN
df.head(9)
```

```python
attr_df = df[df['Attrition']=='yes']
```

```python
# First, ensure attr_df is correctly filtered for 'Yes' attrition (case-sensitive)
attr_df = df[df['Attrition'] == 'Yes']

attr_df = attr_df.groupby(['Department']).size().reset_index(name='count')
attr_df
```

```python
# CREATE A DONUT CHAT
import plotly.graph_objects as go

fig = go.Figure(
    data = [
        go.Pie(
            labels = attr_df['Department'],
            values = attr_df['count'],
            hole = 0.4,      # donut hole
            textposition = 'inside',
            marker = dict(colors = ['red','green'])
        )
    ]
)

fig.update_layout(title = 'Attrition by Department')
fig.show()

```
