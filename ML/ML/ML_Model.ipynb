APPLYING IT ON INSURANCE DATA


```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
# ML Models ---- >
# (1). Supervised ML Models --- > In this , we will trained our labelled data .
# (a). Descrete --- > LogisticRegression , Decisiontreeclassifier , Randomforestclassifer , naive bayes.
# (b). Continuous --- > LinearRegression , D.T. Regressor , R.F.Regressor
# (2). UNsUPERVISED ML Models ---- > In this , we will trained our unlabelled data .
# KMeans , PCA , Apriori .
```

```python
# (1). Supervised ML Models --->
# (a). LinearRegression
```

```python
# Workflow --- >
# Data Ingestion ---- > Data claning and Preprocessing ---- > Select ML model based on
# target column ---- > Model training --- > Model Prediction ---- > Model Evaluation
```

```python
df = pd.read_csv('/content/insurance - insurance.csv')
```

```python
df.head(2)
```

```python
lb = LabelEncoder()

df['sex'] = lb.fit_transform(df['sex'])
df['smoker'] = lb.fit_transform(df['smoker'])
df['region'] = lb.fit_transform(df['region'])
```

```python
x = df.drop(columns= ['charges'])
y = df['charges']

x_train, x_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size=0.2,
                                                    random_state=58)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
