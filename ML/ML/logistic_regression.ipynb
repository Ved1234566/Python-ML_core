```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\vedan\\OneDrive\\Documents\\Regex\\Python\\ML\\CSV Files\\covid_toy - covid_toy.csv")
df.head(2)
```

```python
df.isnull().sum()
```

```python
df['fever']=df['fever'].fillna(df['fever'].mean())
```

```python
#df.isnull().sum()
```

```python
# Encoding
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['gender']=le.fit_transform(df['gender'])
df['cough']=le.fit_transform(df['cough'])
df['city']=le.fit_transform(df['city'])
df['has_covid']=le.fit_transform(df['has_covid'])
```

```python
x=df.drop(columns=['has_covid'])
y=df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

#### When we had targeted data categorical then we use accuracy_score
#### if numerical then use r2_score

```python
accuracy_score(y_test,y_pred)
```
