```python
# CCA -->coomplete case analysis
# df --> missing values  --> fitting missing values-->
# New-df = filter column in which we have missing data
# new-df --> missing data drop

# Architecture --> Histogram -->
# pas column in which we have missing data
# update colun in which we have no missing data

# if past data architecture is overlap to new data architcture
# it means we can drop missing data
# if past data architecture is not overlap to new data architecture
# it means we can drop missing data we will fill missing data
```

```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('/content/dsjob - dsjob.csv')
```

```python
df.isnull().mean()*100
```

```python
cols = [var for var in df.columns if df[var].isnull().mean()<0.05 and df[var].isnull().mean() > 0]
cols
```

```python
df[cols].sample(5)
```

```python
df['education_level'].value_counts()
```

```python
len(df[cols].dropna()) / len(df)
```

```python
new_df = df[cols].dropna()
df.shape , new_df.shape
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# Original Data
df['experience'].hist(bins = 50, ax = ax , density=True , color = 'red')

# data after cca , the argument alpha makes the color transparent so we can
# see the overlay of the 2 distribution
new_df['experience'].hist(bins = 50, ax = ax , density=True , color = 'blue' , alpha = 0.8)

# plt.show()
```
